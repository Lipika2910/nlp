### <h1> PRACTICAL 1 <h1>
Prepare/Pre-process a text corpus to make it more usable for NLP tasks usmg tokenization, conversion to lowercase, removal of punctuation, filtration of stop words, stemming and lemmatization.

```
#import libraries
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')

import spacy
!pyhton -m spacy download en_core_web_sm
```

```
#input Corpus
corpus = """
Natural Language Processing (NLP) is one of the most important fields in the field of artificial intelligence.
It helps computers to understand, interpret, and generate human language. unhappy rewriting Unbelievable"""
```

```
#Tokenization
from nltk.tokenize import word_tokenize
tokens = word_tokenize(corpus)
print(tokens)
```

```
#covert to lowercase and remove punctuations
#The NLTK stopword list is in lowercase. Ensure you convert your text to lowercase before filtering.
tokens_lower = [token.lower() for token in tokens if token.isalpha()]
print("After lowercase and punctuation removal :\n",tokens_lower)
```

```
#Stopword Removal
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens_lower if token not in stopwords]
print("After stopword removal :\n",filtered_tokens)
```

```
#Stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]
print("After Stemming :\n", stemmed_tokens)
```

```
#Lemmatization using spaCy
nlp = spacy.load('en_core_web_sm')
doc = nlp(' '.join(filtered_tokens))
lemmatized_tokens = [token.lemma_ for token in doc]
print("After Lemmatization :\n",lemmatized_tokens)
```

```
#summary of Tokens
print("Tokens in Original : \n",tokens)
print("\nTokens after Lowercase and Punctuation Removal : \n",tokens_lower)
print("\nTokens after Stopword Removal : \n", filtered_tokens)
print("\nTokens after Stemming : \n", stemmed_tokens)
print("\nTokens after Preprocessing : \n", lemmatized_tokens)
```

```
#compare number of tokens before and aftyer preprocessing
print("Number of Tokens in Original : ", len(tokens))
print("Number of Tokens after Lowercase and Punctuation Removal : ", len(tokens_lower))
print("Number of Tokens after Stopword Removal : ", len(filtered_tokens))
print("Number of Tokens after Stemming : ", len(stemmed_tokens))
print("Number of Tokens after Preprocessing : ", len(lemmatized_tokens))
```

Create the TF-IDF (Term Frequency -Inverse Document Frequency) Matrix for the given set of text documents

```
# Import necessary libraries
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize
import re
import numpy as np
import pandas as pd # Ensure pandas is imported for DataFrame operations

# Download NLTK data if not already present
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Add this line to download the missing resource

# Load the corpus from gutenberg and preprocess it as in 'PRACTICE OF PRACTICAL 3'
text = gutenberg.raw('austen-emma.txt')
sentences = sent_tokenize(text)
corpus = sentences[:10]

# Initialize tokenizer and stopwords
wpt = nltk.WordPunctTokenizer()
stop_words = nltk.corpus.stopwords.words('english')

# Define the document normalization function
def normalize_document(doc):
    doc = re.sub(r'[^a-zA-Z\s]', "", doc, re.I | re.A)
    doc = doc.lower()
    doc = doc.strip()
    tokens = wpt.tokenize(doc)
    filtered_tokens = [token for token in tokens if token not in stop_words]
    doc = ' '.join(filtered_tokens)
    return doc

# Normalize the corpus
normalize_corpus = np.vectorize(normalize_document)
norm_corpus = normalize_corpus(corpus)

# Now, proceed with TF-IDF vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(min_df=0., max_df=1., use_idf=True)
tv_matrix = tv.fit_transform(norm_corpus)

# View non-zero features positions in the sparse matrix
print("Sparse TF-IDF Matrix:\n", tv_matrix)

# Get all unique words in the corpus
vocab = tv.get_feature_names_out()
print("\nVocabulary:\n", vocab)

# View dense representation
tv_matrix = tv_matrix.toarray()
print("\nDense TF-IDF Matrix:\n", tv_matrix)

# Show document feature vector as a DataFrame
df = pd.DataFrame(tv_matrix, columns=vocab)
display(df)

# Compute average TF-IDF score for each word
avg_tfidf = df.mean(axis=0)

# Convert to DataFrame
avg_tfidf_df = avg_tfidf.reset_index()
avg_tfidf_df.columns = ['Word', 'Avg_TF_IDF']

# Sort by TF-IDF score
sorted_tfidf = avg_tfidf_df.sort_values(by='Avg_TF_IDF', ascending=False)

# Top 5 words
print("\nTOP 5 WORDS BY TF-IDF SCORE:\n")
print(sorted_tfidf.head(5))

# Bottom 5 words
print("\nBOTTOM 5 WORDS BY TF-IDF SCORE:\n")
print(sorted_tfidf.tail(5))
```

```
# ==============================
# 1. IMPORT LIBRARIES
# ==============================
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize
import re
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Download NLTK data
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('stopwords')

# ==============================
# 2. LOAD & NORMALIZE CORPUS
# ==============================
text = gutenberg.raw('austen-emma.txt')
sentences = sent_tokenize(text)
corpus = sentences[:10]  # First 10 sentences as 10 documents

# Normalization
wpt = nltk.WordPunctTokenizer()
stop_words = nltk.corpus.stopwords.words('english')

def normalize_document(doc):
    doc = re.sub(r'[^a-zA-Z\s]', "", doc)
    doc = doc.lower().strip()
    tokens = wpt.tokenize(doc)
    filtered_tokens = [token for token in tokens if token not in stop_words]
    return ' '.join(filtered_tokens)

norm_corpus = [normalize_document(doc) for doc in corpus]

# ==============================
# 3. TERM FREQUENCY (TF)
# ==============================
count_vectorizer = CountVectorizer()
tf_matrix = count_vectorizer.fit_transform(norm_corpus)
tf_df = pd.DataFrame(tf_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())

print("\n========== TERM FREQUENCY (TF) ==========\n")
display(tf_df)

# ==============================
# 4. DOCUMENT FREQUENCY (DF) & IDF
# ==============================
df_series = (tf_df > 0).sum()  # Count in how many docs each word appears
idf_values = np.log((len(tf_df) + 1) / (df_series + 1)) + 1  # Standard IDF formula

idf_df = pd.DataFrame({
    'Word': df_series.index,
    'Document_Frequency': df_series.values,
    'IDF': idf_values.values
}).sort_values(by='IDF', ascending=False)

print("\n========== DOCUMENT FREQUENCY (DF) & IDF ==========\n")
display(idf_df)

# ==============================
# 5. TF-IDF MATRIX
# ==============================
tfidf_matrix = tf_df * idf_values.values  # Element-wise multiplication
tfidf_df = pd.DataFrame(tfidf_matrix.values, columns=tf_df.columns)

print("\n========== TF-IDF MATRIX ==========\n")
display(tfidf_df)

# ==============================
# 6. AVERAGE TF-IDF
# ==============================
avg_tfidf = tfidf_df.mean().sort_values(ascending=False)
avg_tfidf_df = pd.DataFrame({
    'Word': avg_tfidf.index,
    'Avg_TF_IDF': avg_tfidf.values
})

print("\n========== AVERAGE TF-IDF ==========\n")
display(avg_tfidf_df)

# ==============================
# 7. OPTIONAL: Top & Bottom 5
# ==============================
print("\nTop 5 words by Avg TF-IDF:\n")
display(avg_tfidf_df.head(5))

print("\nBottom 5 words by Avg TF-IDF:\n")
display(avg_tfidf_df.tail(5))
```

### <h1> PRACTICAL 5<H1/>
5. Build a simple statistical language model: estimate unigram and bigram probabilities with add‑one smoothing and compute the probability of given sentences.

```
#define corpus
import nltk
nltk.download('punkt') # Download the 'punkt' tokenizer
from nltk.tokenize import word_tokenize
from nltk.util import bigrams, ngrams # Import ngrams

# Sample text
sentences_list = [ "I am Sam",
       "Sam I am",
       "I do not like green eggs and ham"
]

# Join the list of sentences into a single string
text_string = ' '.join(sentences_list)
text_string = text_string.lower()
text_string
```

```
#Tokenize and print unigram and bigram

import nltk
nltk.download('punkt_tab')
from nltk.tokenize import RegexpTokenizer # Import RegexpTokenizer

# Tokenize the text string
tokens = word_tokenize(text_string)

#unigram list
unigram_list = list(ngrams(tokens, 1)) #use ngrams with n=1

print('\nUnigrams:')
for unigram in unigram_list:
    print(unigram)


tokenizer = RegexpTokenizer(r'</?s>|\w+|[^\w\s]')

# Add sentence markers
sentences_with_markers = [f"<s> {sentence.lower()} </s>" for sentence in sentences_list]

# Join into one string
bigram_text_string = ' '.join(sentences_with_markers)

# Tokenize
bitokens = tokenizer.tokenize(bigram_text_string)
unitokens = tokenizer.tokenize(bigram_text_string)

# Generate bigrams
bigram_list = list(bigrams(bitokens))
bi_unigram_list = list(ngrams(unitokens, 1))

print("\nBigrams:")
for bg in bigram_list:
    print(bg)
```

```
#count unigram frequency
from collections import Counter

unigram_counts = Counter(unigram_list)
print(unigram_counts)
```

```
#count bigram frequency
bi_unigram_counts = Counter(bi_unigram_list)
print(bi_unigram_counts)
bigram_counts = Counter(bigram_list)
print(bigram_counts)
```

```
#count total number of words for unigram
total_tokens = len(tokens)
print(total_tokens)
```

```
#calculate unigram probabilities
p_I = unigram_counts[('i',)] / total_tokens
print("P(i) =", p_I)
```

```
#calculate all unigram probabilities
print("\nUNIGRAM PROBABILITIES\n")

for unigram, count in unigram_counts.items():
    prob = count / total_tokens
    print(f"P({unigram[0]}) = {count}/{total_tokens} = {prob:.4f}")
```

```
#calculate bigram probabilities
p_am_given_I = bigram_counts[('i', 'am')] / unigram_counts[('i',)]
print("P(am | i) =", p_am_given_I)
```

```
#calculate all bigram probabilities
print("\nBIGRAM PROBABILITIES\n")

for bigram, bigram_count in bigram_counts.items():
    w1, w2 = bigram
    if w2 != '<s>' and w1 != '</s>':
      unigram_count = bi_unigram_counts[(w1,)]
      prob = bigram_count / unigram_count
      print(f"P({w2} | {w1}) = {bigram_count}/{unigram_count} = {prob:.4f}")
```

```
#Print ALL UNIGRAM probabilities (smoothed)
total_tokens = len(tokens)
vocab_size = len(unigram_counts)

print("\nUNIGRAM PROBABILITIES (ADD-ONE SMOOTHING)\n")

for unigram, count in unigram_counts.items():
    prob = (count + 1) / (total_tokens + vocab_size)
    print(f"P({unigram[0]}) = ({count}+1)/({total_tokens}+{vocab_size}) = {prob:.4f}")
```

```
#Print ALL BIGRAM probabilities (smoothed)
vocab_size = len(unigram_counts)

print("\nBIGRAM PROBABILITIES (ADD-ONE SMOOTHING)\n")

for bigram, bigram_count in bigram_counts.items():
    w1, w2 = bigram
    if w2 != '<s>' and w1 != '</s>':
      unigram_count = bi_unigram_counts[(w1,)]
      prob = (bigram_count + 1) / (unigram_count + vocab_size)
      print(f"P({w2} | {w1}) = ({bigram_count}+1)/({unigram_count}+{vocab_size}) = {prob:.4f}")
```

```
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GlobalAveragePooling1D, Dense

# Load dataset
dataset, info = tfds.load("ag_news_subset", with_info=True, as_supervised=True)

train_data, test_data = dataset['train'], dataset['test']

X_train, y_train = [], []
X_test, y_test = [], []

for text, label in train_data:
    X_train.append(text.numpy().decode("utf-8"))
    y_train.append(label.numpy())

for text, label in test_data:
    X_test.append(text.numpy().decode("utf-8"))
    y_test.append(label.numpy())
```

```
print("DATASET OVERVIEW")

train_size = len(X_train)
test_size = len(X_test)
total_size = train_size + test_size

print("\n Total dataset size:", total_size)
print("Train size:", train_size)
print("Test size:", test_size)

label_names = info.features['label'].names
num_labels = len(label_names)

print("\n Number of labels:", num_labels)

print("\n Label names:")
for i, name in enumerate(label_names):
    print(f"{i} → {name}")

print("\n Sample News Data (5 examples):\n")

for i in range(5):
    print(f"Sample {i+1}")
    print("News  :", X_train[i])
    print("Label :", label_names[y_train[i]])
    print("-" * 80)
```

```
max_words = 10000
max_len = 100

tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post')
```

```
history = model.fit(
    X_train_pad,
    np.array(y_train),
    validation_data=(X_test_pad, np.array(y_test)),
    epochs=3,
    batch_size=64
)
```

```
# Create Sequences

seq_length = 3

X = []
y = []

for i in range(0, len(text) - seq_length, 3):
    seq = text[i:i+seq_length]
    label = text[i+seq_length]

    X.append([char_to_idx[c] for c in seq])
    y.append(char_to_idx[label])

X = np.array(X)
y = np.array(y)

print("Total samples:", X.shape[0])
```

```
# sampling fucnction

def sample(preds, temperature=0.7):
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds + 1e-8) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    return np.random.choice(len(preds), p=preds)
```

```
# prediction function

def predict_next_char(text_input):
    # FIX: pad input if smaller than seq_length
    if len(text_input) < seq_length:
        text_input = " " * (seq_length - len(text_input)) + text_input

    text_input = text_input[-seq_length:]

    seq = [char_to_idx.get(c, 0) for c in text_input]
    seq = np.reshape(seq, (1, seq_length, 1)) / float(vocab_size)

    pred = model.predict(seq, verbose=0)[0]

    return idx_to_char[np.argmax(pred)]
```

```
# generate text


def generate_text(seed_text, length=300, temperature=0.7):
    # pad if needed
    if len(seed_text) < seq_length:
        seed_text = " " * (seq_length - len(seed_text)) + seed_text

    result = seed_text

    for _ in range(length):
        seq = [char_to_idx.get(c, 0) for c in seed_text[-seq_length:]]
        seq = np.reshape(seq, (1, seq_length, 1)) / float(vocab_size)

        preds = model.predict(seq, verbose=0)[0]
        next_index = sample(preds, temperature)

        next_char = idx_to_char[next_index]

        result += next_char
        seed_text += next_char

    return result
```
